# What is Customer Lifetime Value (CLTV)?

Customer Lifetime Value is one of the key metrics in defining the amount of revenue we can expect from a customer. This metric is useful in that it is important for us keep track of how much a customer spends in a lifetime and how much revenue we can expect from a customer during the customer lifetime. In an ecommerce business our ideal customers are those that continue to purchase on a regular basis ensuring that the company has a steady flow of revenue. There are many ways to calculate customer lifetime value depending on the type of business and the customer base. Below we will go over one of the ways to calculate customer lifetime value based on an ecommerce retail data set.

Some questions that it can help a business to answer are:

    How to Identify the most profitable customers?

    How can a company offer the best product and make the most money?

    How to segment profitable customers?

    How much budget need to spend to acquire customers?

### What is the formula for calculating CLTV?

The actual formula for calculating CLTV will vary depending on the business, but we will use an appropriate CLTV formula for our ecommerce data set.

CLTV = ((Average Order Value * Purchase Frequency)/Churn Rate) * Profit Margin

### Here we will calculate three things:
    Calculate number of days between present day and last purchase date for each customer
    Calculate the number of orders per customer
    Calculate the total item quantity per customer
    Calculate the total order value per orders

```python
uk_df_group = uk_df.groupby('CustomerID').agg({'InvoiceDate': lambda date: (date.max() - date.min()).days,
                                        'InvoiceNo': lambda num: len(num),
                                        'Quantity': lambda qty: qty.sum(),
                                        'TotalOrderValue': lambda price: price.sum()})
```

```python
# Change the name of columns
uk_df_group.columns=['num_days','num_transactions','num_units','spent_money']
uk_df_group.head()
```

# Calculate CLTV using the following formula:

The actual formula for calculating CLTV will vary depending on the business, but we will use an appropriate CLTV formula for our ecommerce data set.

 CLTV = ((Average Order Value x Purchase Frequency)/Churn Rate) x Profit margin.

 Customer Value = Average Order Value * Purchase Frequency


### Calculate Average Order Value

The average order value is the total revenue of a customer divided by the number of orders placed within a set time period.

Average Order Value = (Total Revenue from Customer / Total Number of Orders)

```python
uk_df_group['avg_order_value'] = uk_df_group['spent_money'] / uk_df_group['num_transactions']
uk_df_group['avg_order_value']
```

### Calculate Purchase Frequency

The purchase frequency is defined as the ratio of total number of orders divided by the total number of customers. 

Purchase Frequency = (Total Number of Orders / Total Number of Customers)

```python
purchase_frequency = sum(uk_df_group['num_transactions']) / uk_df_group.shape[0]
purchase_frequency
```

### Calculate Repeat Rate

Repeat rate can be defined as the number of customers that return to purchase. It is the ratio of number of customers with more than one order to the total number of unique customers.

Repeat Rate = (Number of Customers with greater than 1 order) / Total Number of Unique Customers

```python
repeat_rate = uk_df_group[uk_df_group['num_transactions'] > 1].shape[0] / uk_df_group['num_transactions'].shape[0]
repeat_rate
```

### Calculate Churn Rate

The churn rate is the number of customers who have not purchased again. The churn rate will vary depending on the business and will alter the CLTV value based on when a customer is considered to be churned. It is also inversely related to the repeat rate.

Churn Rate = 1 - Repeat Rate

```python
churn_rate = 1 - repeat_rate
churn_rate
```

### Calculate Profit Margin

Using the Profit Margin we will assume that the net profit margin for an ecommerce business is ~ 15%

```python
uk_df_group['profit_margin'] = uk_df_group['spent_money']*.15
```

### Calculate Customer Lifetime Value

Customer lifetime is the period of time that the customer has been purchasing

Customer Lifetime = 1/Churn Rate

```python
# Customer Value
uk_df_group['CLV'] = (uk_df_group['avg_order_value']*purchase_frequency) / churn_rate
```

```python
# Customer Lifetime Value
uk_df_group['cust_lifetime_value'] = uk_df_group['CLV']*uk_df_group['profit_margin']
```

# Prediction Model for CLTV

```python
uk_df.head()
```

Extract month and year from Invoice Date

```python
uk_df['month_yr'] = uk_df['InvoiceDate'].apply(lambda x: x.strftime('%b-%Y'))
```

Create two-dimensional table view of summarized data

```python
uk_sales = uk_df.pivot_table(index=['CustomerID'],columns=['month_yr'],values='TotalOrderValue',aggfunc='sum',fill_value=0).reset_index()
uk_sales.head()
```

```python
# select all rows and starting from 3rd column and sum along rows to get the 'CLV'

uk_sales['CLV'] = uk_sales.iloc[:,2:].sum(axis=1)  #iloc[start row: end row(incl), start col: end col(incl)]
```

### Selecting Features

```python
X = uk_sales[['Dec-2011','Nov-2011', 'Oct-2011','Sep-2011','Aug-2011','Jul-2011']]
y = uk_sales[['CLV']]
```

### Train Test Split Dataset

```python
#split training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### Model Development

```python
# import model
from sklearn.linear_model import LinearRegression

# instantiate
linreg = LinearRegression()

# fit the model to the training data (learn the coefficients)
linreg.fit(X_train, y_train)

# make predictions on the testing set
y_pred = linreg.predict(X_test)
```

```python
# print the intercept and coefficients
print(linreg.intercept_)
print(linreg.coef_)
```

# How Does the Model Fit the Data?

What is R squared?
R-squared is a statistical measure that represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model. A model where R-squared value is 70% would mean that the model explains 70% of the fitted data in the regression model.




```python
from sklearn import metrics

# compute the R Square for model
print("R-Square:",metrics.r2_score(y_test, y_pred))
```

What is Mean Absolute Error?
Absolute Error is the difference between the measured value and "true" value. The Mean Absolute Error (MSE) with respect to the test set is the mean of the absolute values of the individual prediction errors over all instances of the test set. The lower the MAE, the better a model fits a dataset.

What is Mean Squared Error?
Mean Square Error (MSE) corresponds to the average of the square of the errors. The larger the number the larger the error. An ideal MSE value is 0.0, which means that all the predicted values matched the expected values exactly.

What is Root Mean Square Error?
Root Mean Square Error (RMSE) is defined as the standard deviation of the residuals (prediction errors). Residuals are a measure of how far from the regression line data paoints are. It tells yo how concentrated the data is around the line of best fit. Based on a rule of thumb, it can be said that RMSE values between 0.2 and 0.5 shows that the model can relatively predict the data accurately. The lower the RMSE, the better a given model is able to "fit" a dataset.


```python
# calculate MAE using scikit-learn
print("MAE:",metrics.mean_absolute_error(y_test,y_pred))

#calculate mean squared error
print("MSE",metrics.mean_squared_error(y_test, y_pred))

# compute the RMSE of our predictions
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
