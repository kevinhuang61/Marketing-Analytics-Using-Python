# Exploratory Data Analysis

Get the distribution of the purchase count using a boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

users = data['CustomerID'].value_counts().reset_index()
users.columns = ['CustomerID', 'Purchase Count']
sns.boxplot(users['Purchase Count'], sym='')
plt.show()
```

We need to isolate the time value of 'InvoiceDate' to their respective date time values using a lambda function converting each unit to UTC time

```python
import dateutil
from datetime import datetime as dt
from pytz import utc

data['datetime'] = data['InvoiceDate'].apply(lambda x: dateutil.parser.parse(x).timestamp())
data['month'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).month)
data['day'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).day)
data['year'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).year)
data['hour'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).hour)
data['minute'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).minute)
print(data)
```

Dropping values NA values along the CustomerID column since we can not identify those customers as new or returning users and if or when they re-engage with us.

```python
data_notna = data.dropna(axis=0)
data_notna.info()
```

Create a new dataframe to store cohorts based on the month and year that they first purchased from the store, grouping by CustomerID and merging with data_notna Dataframe

```python
data_notna['cohort'] = data_notna.apply(lambda row: (row['year'] * 100) + (row['month']), axis=1)
cohorts = data_notna.groupby('CustomerID')['cohort'].min().reset_index()
print(cohorts)
cohorts.columns = ['CustomerID', 'first_cohort']
data_notna = data_notna.merge(cohorts, on='CustomerID', how='left')
```

# Graphing the cohort analysis using heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig_dims = (12, 8)
fig, ax = plt.subplots(figsize=fig_dims)
y_labels = [str(int(header)%100) + '-' + str(int(header)/100) for header in headers]
x_labels = range(0, len(y_labels))
plt.yticks(ticks=headers, labels=y_labels)
plt.xticks(x_labels, x_labels)
ax.set(xlabel='Months After First Purchase', ylabel='First Purchase Cohort', title="Cohort Analysis")
sns.heatmap(cohort_pivot, annot=True, fmt='.0%', mask=cohort_pivot.isnull(), ax=ax, square=True, linewidths=.5, cmap=sns.cubehelix_palette(8))

plt.show()
```

# Recommendations

We can compare user retention rates among different cohorts to see if there is an improvement in retention rates over time. 

Doing this type of analysis in Python allows one to combine data from different sources to see how customer cohorts change over time and can allow us to do more ad-hoc analyses compared to out-of-the-box analytics platforms like Google Analytics.

An interesting statistic is that 11 months after the initial cohort from the 201012 cohort, 50% of those users returned to make an additional purchase. We might want to look into why we have such a high re-engagement rate almost one year after the initial purchase.

For further analysis we can apply this analysis to different customer segments or to different customers that received different marketing campaigns to see how user retention varies. This will let us know if a particular marketing campaign works the best, or if we should target a specific customer segment based on retention rates.
